```python
X = train.drop(columns = ['Response', 'id'])
y = train['Response']
```

```python
X[:5]
```

```python
cat_features = ['Gender','Vehicle_Age','Vehicle_Damage','Driving_License','Previously_Insured']
num_features = ['Age','Region_Code','Annual_Premium','Policy_Sales_Channel','Vintage']
```

```python
x_train, x_val, y_train, y_val = train_test_split(X,y,test_size=0.3, random_state = 24)
```

```python
cat_transformer = Pipeline(steps=[('onehot',OneHotEncoder())])
```

```python
num_transformer = Pipeline(steps=[('scale',StandardScaler())])
```

```python
preprocessor = ColumnTransformer (transformers = [('categorical_transformer',cat_transformer,cat_features),
                                                  ('Numerical_transformer',num_transformer,num_features)])
```

```python
model_pipeline = Pipeline(steps=[
    ('Preprocessing', preprocessor), 
    ('Logistic_regression_model', LogisticRegression())
])
```

```python
model_pipeline
```

```python
model_pipeline.fit(x_train,y_train)
```

```python
train_pred = model_pipeline.predict(x_train)
```

```python
val_pred = model_pipeline.predict(x_val)
```

```python
accuracy_score(y_train, train_pred)
```

```python
accuracy_score(y_val, val_pred)
```
